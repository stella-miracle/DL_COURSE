```python
#  Import necessary libraries
!pip install tensorflow

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Dense
import warnings
warnings.filterwarnings('ignore',category=UserWarning,module='tensorflow')
```

```python
# define input layer
input_layer = Input(shape=(20,))
# creates an input layer that expects input vector of length 20.
print(input_layer)
```

```python
# add hidden layer
hidden_layer1 = Dense(64,activation = 'relu')(input_layer)
# Dense(fully connected )
# each hidden layer takes the output of the previous layer as input.
hidden_layer2 = Dense(64,activation = "relu")(hidden_layer1)
```

```python
# define output layer
output_layer = Dense(1, activation="sigmoid")(hidden_layer2)
# sigmoid activate function is suitable binary classicifaction.
```

```python
# create the model
model = Model(inputs = input_layer, outputs = output_layer)
model.summary()
```

```python
#complie the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
# optimizer= a popular choice for training neural network
# binary_crossentropy = specifies the loss function for binary classification problems.
# metrics = instructs keras to evalute the model using accuracy during training.
```

```python
# train the model
import numpy as np
x_train = np.random.rand(1000,20)
# 0-1 사이의 실수를 1개만 생성
# create a random number (0~1)

# random.rand = generate random numbers following a uniform distribution
# The second demand is 0, meaning x_train will have 100 rows but only 0 columns.
y_train = np.random.randint(2, size = (1000,1))
# 0이상 2미만, (1000,1) 사이즈의 '정수' 생성

model.fit(x_train, y_train, epochs=10, batch_size = 31)
```

```python
# evalutate the model
X_test = np.random.rand(200,20)
# predictoin value after x_train value, have to set the same feature number as x_train,
# but the data will be new one,

Y_test = np.random.randint(2,size=(200,1))
# y_test is right answer of x_test.

loss, accuracy = model.evaluate(X_test,Y_test)
print(f'test loss:{loss}')
print(f'accracy : {accuracy}')
```

####Dropout layer?


It's a regularzation techniques that help prevents overfitting in neural network.
During traning, dropout randomly sets a fraction of input units to zero at each update cycle.

"(훈련중 dropout 은 매 업데이트 주기마다 입력유닛의 일부를 무작위로 0 으로 설정한다. )"

prevents the model from becoming overly reliant on any specific neurons, which encourage the network to learn more robust features and generalize better to unseen data.

***key points***

- Dropout is only appiled during training, not during inference. "(추론(예측)"
-The dropout rate is a hyperparameter that determines the fraction of neurnos to drop.

####Batch Normalization?

it's a techniqueu used to improve the training stability and speed of neural networks. it normalizes the output of the previous layer by re-centreing and re-scaling the data, which helps in stabilizing the learning process. also prevent oeverfitting.


***key points***

- Batch normalization works by normalizing the inputs to each layer to have  mean of zero and a variance of one."(배치 정규화는 각 층의 입력을 정규화하여 평균이 0 이고 분산이 1 이 되도록 한다)"
- it's applied during both training and inference, although its behavior varies slightly between the two phases."훈련과 예측 과정 모두 사용가능하지만, 두 단계간 동작은 약간씩 다르다."
- Batch normalization layers also introduce two learable parameters that allow the model to scale and shift the normalization output, which helps in restoring the model's representational power.


```python
# e.g dropout layer in keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout, Dense, Input

input=Input(shape=(20,0))
hidden1 = Dense(64, activation='relu')(input_layer)
# dropout
drop = Dropout(rate = 0.5)(hidden1)
hidden2 = Dense(64, activation='relu')(drop)
output= Dense(1, activation= 'sigmoid')(hidden2)
model = Model(inputs = input, outputs = output)
model.summary()
```
