```python
# import nesssary libraries

!pip install tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Layer
from tensorflow.keras.layers import Softmax

# make custom dense layer with 32 units and relu activatoin
# define custom layer

class CustomDenseLayer(Layer):
    def __init__(self, units=32):
        super(CustomDenseLayer, self).__init__()
        self.units = units

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(input_shape[-1], self.units),
            initializer="random_normal",
            trainable=True
        )
        self.b = self.add_weight(
            shape=(self.units,),
            initializer="random_normal",
            trainable=True
        )

    def call(self, inputs):
        return tf.nn.relu(tf.matmul(inputs, self.w) + self.b)
        # relu = positive -> same value, negative -> 0

#Define the model with sofrmax in the outout layer
model = Sequential([
    CustomDenseLayer(128),
    CustomDenseLayer(10), #hidden layer with relu activatoin
    Softmax()      # outpur layer with softmax activatoin for mulit-class classificatoin
])


# compile the model
model.compile(optimizer='adam', loss ='categorical_crossentropy', metrics=['accuracy'])
print('model summary before building:')
model.summary()

#build model
model.build((1000,20))
print("\n model summary after building:")
model.summary()
```

```python

# train the model
import numpy as np
X_train= np.random.random((1000,20))
# random.random -> single random number (float), not possible for dimension specification.
# random.rand -> multidimentional array(ndarray), possible for dimenesion specificatoin.
y_train = np.random.randint(10, size=(1000,1))
# create 0-9 random int nunmber, 1000 labels valut, 1 column.

# convert labers to categorical one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train,num_classes=10)
model.fit(X_train, y_train,epochs=10, batch_size=32)
```

```python
#evaluate the model

x_test = np.random.random((200,20))
y_test = np.random.randint(10, size =(200,1))

# convert label
y_test = tf.keras.utils.to_categorical(y_test,num_classes=10)

# evaluate model
loss,accuracy = model.evaluate(x_test, y_test)
print(f'loss:{loss}')
print(f"accuracy:{accuracy}")
# if you don't put'metrics' parameter in model.compile, you can't see accuracy value in here.
# it has to get error named "type error: cannot unpack non-iterable float object"
```
