```python
import numpy as np
import pandas as pd

weight = np.around(np.random.uniform(size=6),decimals=2)
# around : 실수를 소수점 기준반올림, decimals =? 몇번째자리까지 반올림할건지
# random.uniform() : 주어진 구간 (low, high) 에서 균일하게 분포된 난수 설정.
# random.uniform(low, high, size) -> 난수의 최솟값(포함), 최댓값(미포함), 생성할 난수의 개수 또는 배열 차원

biases = np.around(np.random.uniform(size=3),decimals=2)
```

```python
print(weight)
print(biases)
```

```python
# input setting
x1 = 0.5
x2 = 0.85

print("x1 is {} and x2 is {}".format(x1,x2))
```

```python
z1=x1*weight[0]+x2*weight[1]+biases[0]
```

```python
print("z1 is {}".format(z1))
```

```python
z1_2=x1*weight[2]+x2*weight[3]+biases[1]
print("z1_2 is {}".format(z1_2))
```

```python
a1 = 1.0/(1.0+np.exp(-z1))
```

```python
a1_2 = 1.0/(1.0+np.exp(-z1_2))
```

```python
z2 = a1*weight[4] + a1_2*weight[5]+biases[2]
```
