```python
# import keras
from keras.models import Sequential
from keras.layers import Input
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
 # conventional nerual network 로 갈 경우 -> we need extra package
 from keras.layers import Conv2D #to add conventional layer
 from keras.layers import MaxPooling2D #to add pooling layer
 from keras.layers import Flatten # to flatten


```

```python
#import data
from keras.datasets import mnist
```

```python
#load data
(x_train, y_train),(x_test,y_test) = mnist.load_data()
```

```python
# 데이터 전처리과정-> 각 열의 데이터를 표준화 하여 모델 학습 시 더 안정하고 빠르게 학습시키기 위한 방법
# 그전에 했던 regrssion 이나 classification 에서는 1차원 벡터(한 이미지의 총 픽셀수 )로 평탄화 했기 때문에
# num_pix = x.train.shape[1]*x_train.shape[2] 곱한 값을 통해 1차원에서 784차원의 백터로 변경 후
# 데이터 flatten 작업 수행 시 -> x_train= x_train.reshape(x_train[0], num_pic).astype("float32") 로 샘플수와 픽셀수로 계산 - > dense 레이어에 넣기위해 사용
```

```python
# CNN 에서 데이터 전처리 과정은 위의 과정과 다른 부분이 존재 -> 입체적인 형태를 유지 즉, 이미지를 일차원의 픽셀값이 아닌 2,3 차원의 형태로 유지할때 사용하는 전처리 방법
# 이미지를 (샘플수, 높이 , 너비, 채널 수 ) 로 만든다.
# MNIST 데이터셋이 28*28 크기의 흑백 이미지 였다면, 각이미지는 (28*28*1) 의 형태가 된다.
# 이런 방식의 전처리는 CNN(합성곱 신경망) 에서 이미지의 공간적 구조를 그대로 유지하며 사용한다

x_train = x_train.reshape(x_train.shape[0],28,28,1).astype('float32')
x_test= x_test.reshape(x_test.shape[0],28,28,1).astype('float32')
```

```python
# 0-1 까지 데이터 전처리 과정 -> 이 과정은 같다
x_train=x_train/255
x_test= x_test/255
```

```python
y_train=to_categorical(y_train)
y_test =to_categorical(y_test)

num_class= y_test.shape[1]
```

```python
num_class
```

```python
# make function
def conventional_model():
    model = Sequential()
    model.add(Input(shape=(28,28,1)))  # Corrected 'input' to 'Input'
    model.add(Conv2D(16, (5, 5), strides=(1, 1), activation="relu"))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))  # Corrected 'addr' to 'add'

    model.add(Flatten())
    model.add(Dense(100, activation="relu"))
    model.add(Dense(num_class, activation="softmax"))  # num_class is passed as a parameter

    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model
```

```python
# build model
model=conventional_model()
```

```python
# fit the model
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=200, verbose=2)
```

```python
# evaluate model
score = model.evaluate(x_test, y_test, verbose=0)
# verbose -> 진행상황 표현하지 않기 위한 코드
print("accuracy:{}\n, error{}%".format(score[1],100-score[1]*100))
# score[1]-score[1]*100 -> 정확도(%)를 100에서 뺀값이므로 score[1]에 *100 을 해서 확률로 나오게함.
```
